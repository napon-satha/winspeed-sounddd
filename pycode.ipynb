```python
import pandas as pd

df = pd.read_csv('/Users/naponsathavirawong/Downloads/Customer_all.csv', na_values=['NA', 'missing'])
```

```python
df
```

Cleaned23to24 = 
Filter between 23-01-01 to 24-11-12
Sort DESC invDate 
Unique CustID

```python

cleaned23to24 = df[(df['InvDate'] >= '2023-01-01') & (df['InvDate'] <= '2024-11-12')]
```

```python
cleaned23to24 = cleaned23to24.sort_values('InvDate',ascending=False).drop_duplicates(subset=['CustID'])
cleaned23to24
```

clean_existuser =
Filter Oldest to 22-12-31
Filter Exist_user = 1
Unique Cust_Id

```python
clean_existuser = df[(df['InvDate'] <= '2022-12-31') & (df['Exist_user'] == 1)]
clean_existuser = clean_existuser.drop_duplicates(subset=['CustID']).sort_values('CustID')
```

```python
clean_existuser
```

combinedata1 = 
Concat clean_existuser และ cleaned23to24

```python
combined_df = pd.concat([clean_existuser, cleaned23to24], ignore_index=True)
combined_df
```

```python
combined_df = combined_df.sort_values('InvDate',ascending=False).drop_duplicates(subset=['CustID'])
```

```python
pattern = r'(?i)^(พระ|หสม\.|มหาวิทยาลัย|วิทยาลัย|สำนัก|หจก\.|ธนาคาร|วัด|mr\.|mrs\.|k\.|หจก\.|ห้างหุ้นส่วนจำกัด|ด\.ต\.|บจก\.|ส\.ต\.ท\.|สำนักงาน|พล\.ต\.ต\.|พล\.ต\.ท\.|พล\.ต\.อ\.|พลโท|พลตรี|คุณ|นาย|นาง|นางสาว|mr|mrs|ms|หสม|หจก|บจก|ธนาคาร|โรงเรียน|กิจการร่วมค้า|สถาบัน|สหกรณ์|สภา|โรงพยาบาล|ร้าน|โรงเรียน|วัด|วิทยาลัย|MR\.|Mr\.|K\.|k\.|บริษัท|คุณ|น\.อ\.|น\.ท|ดร\.|ช่าง|คณะ|จ่า|พ\.ต\.ท\.|ร\.ต\.อ\.|คริสตจักร|ครัว|สมาคม|ร\.ต\.|ร\.ต\.ท\.|ร\.ต|รพ\.|มูลนิธิ|องค์การ|สวท\.|ศูนย์|ร\.อ\.|จ\.|กระทรวง|กรม|สว\.|พ\.ต\.|พ\.ต\.อ\.|น\.ต\.|ชมรม|กอง|พล\.ต\.|พ\.อ\.|การไฟฟ้า|เทศบาล|ว่าที่|ภาควิชา)'
```

```python
# สร้าง column title
combined_df['Title'] = combined_df['CustName'].str.extract(pattern).fillna('None')
```
